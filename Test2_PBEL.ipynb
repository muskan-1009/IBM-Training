```python
import nltk
nltk.download('punkt_tab') # Tokenisation
nltk.download('stopwords') # Removal of Stop Words
nltk.download('wordnet') # Lemmatisation

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer # Stemming
from nltk.stem import WordNetLemmatizer
```

Tokenisation

```python
df['tokens'] = df['Message'].apply(word_tokenize)
df['tokens']
```

Stop Word Removal

```python
stop_words = set(stopwords.words("english"))

filtered_tokens = []

for token in df["tokens"]:
    temp = []

    for word in token:
        if word.lower() not in stop_words:
            temp.append(word)

    filtered_tokens.append(temp)

df["Filtered_Tokens"] = filtered_tokens

print(df["Filtered_Tokens"])
```

Stemming

```python
stemmer = PorterStemmer()

stemmed_tokens = []

for token in df["Filtered_Tokens"]:
    temp = []

    for word in token:
      temp.append(PorterStemmer().stem(word))

    stemmed_tokens.append(temp)

df["Stemmed_Tokens"] = stemmed_tokens

df["Stemmed_Tokens"]
```

Lemmatization

```python
lemmatizer = WordNetLemmatizer()

lemmatized_tokens = []

for token in df["Stemmed_Tokens"]:
    temp = []

    for word in token:
        temp.append(lemmatizer.lemmatize(word))

    lemmatized_tokens.append(temp)

df["Lemmatized_Tokens"] = lemmatized_tokens

print(df["Lemmatized_Tokens"])
```

Tokens into Sentences

```python
df["Processed_Message"] = df["Lemmatized_Tokens"].apply(lambda x: " ".join(x))

df[["Processed_Message", "Label"]]
```

Text into Numbers

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(df["Processed_Message"])

y = df["Label"]
```

Splittion of Dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

Training

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train, y_train)
```

Accuracy

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = knn.predict(X_test)

print("Accuracy =", accuracy_score(y_test, y_pred))

print(classification_report(y_test, y_pred))
```

User Message Prediction

```python
msg = input("Enter the message: ")
```

```python
tokens = word_tokenize(msg)

tokens = [word for word in tokens if word.lower() not in stop_words]

tokens = [stemmer.stem(word) for word in tokens]

tokens = [lemmatizer.lemmatize(word) for word in tokens]

processed_message = " ".join(tokens)
```

```python
processed_message
```

```python
message_vector = vectorizer.transform([processed_message])
```

Prediction

```python
prediction = knn.predict(message_vector)

print("Prediction:", prediction[0])

if prediction[0] == "Spam":
    print("This message is SPAM.")
else:
    print("This message is HAM.")
```
