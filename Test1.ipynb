Load Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = [1, 2, 3, 4, 5, None, 7, 8, 9, 10, 11, 12, None, 14, 15, 16, 17, 18, 19, 20]
y_reg = [2, 4, 6, 8, 10, None, 14, 16, 18, 20, 22, 24, None, 28, 30, 32, 34, 36, 38, 40]
y_cls = [0, 0, 0, 0, 0, None, 0, 1, 1, 1, 1, 1, None, 1, 1, 1, 1, 1, 1, 1]
```

```python
df = pd.DataFrame({'X': X, 'y_reg': y_reg, 'y_cls': y_cls})
df
```

Null Values

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
df
```

```python
X = df[['X']]
y_reg = df['y_reg']
y_cls = df['y_cls']
```

Scatter Plot

```python
plt.scatter(X, y_reg)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('y_reg')
plt.show()
```

```python
import sklearn.model_selection as model_selection
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

Linear Regression

```python
linear_model = LinearRegression()
linear_model.fit(X, y_reg)
```

```python
linear_prediction = linear_model.predict([[7]])
linear_prediction
```

Logistic Regression

```python
logistic_model = LogisticRegression()
logistic_model.fit(X, y_cls)
logistic_model
```

```python
logistic_prediction = logistic_model.predict([[4]])
logistic_prediction
```

KNN Classifier

```python
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X, y_cls)
knn_model
```

```python
knn_prediction = knn_model.predict([[5]])
knn_prediction
```

Random Forest Classifier

```python
rf_model = RandomForestClassifier(n_estimators=10, random_state=42)
rf_model.fit(X, y_cls)
rf_model
```

```python
rf_prediction = rf_model.predict([[5]])
rf_prediction
```

Predictions

```python
print("Linear Regression Prediction for X=7 :", linear_prediction[0])
print("Logistic Regression Prediction for X=4 :", logistic_prediction[0])
print("KNN Prediction for X=5 :", knn_prediction[0])
print("Random Forest Prediction for X=5 :", rf_prediction[0])
```
